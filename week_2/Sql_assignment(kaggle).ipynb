# SQL assignment on kaggle dataset  "Sample_Superstore.csv"

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("celebal_internship.db")
```

```python
df = pd.read_csv("Sample_Superstore.csv", encoding="latin1")
```

```python
df.head()
```

## 1.Load dataset into a SQL database

```python
df.to_sql(
    "sales_data",
    conn,
    if_exists="replace",
    index=False
)
```

## 2.Explore table (schema, sample data)

## 3.Apply WHERE filters (region, category, date, sales)

### Filtering data of West region

```python
query = """
SELECT *
FROM sales_data
WHERE Region = 'West'
"""

pd.read_sql(query, conn)
```

The above  query shows all the details of the West region.

### Filtering data with sales>1000

```python
query = """
SELECT
"Order ID" , "Customer Name" , Category, "Product Name" , Sales , Profit
FROM sales_data
WHERE Sales > 1000
ORDER BY Sales DESC;
"""

pd.read_sql(query, conn)
```

This query identifies high-value orders with sales greater than 1000. It helps highlight premium products and customers contributing significantly to revenue.

## 4.Use GROUP BY for aggregations (sales, quantity, averages)

### Orders per Region

```python
query = """
SELECT Region,COUNT(*) AS orders
FROM sales_data
GROUP BY Region
ORDER BY orders DESC
"""

pd.read_sql(query, conn)
```

The above  query shows the total orders in different regions.

### Sales per Region

```python
query = """
SELECT Region,SUM(Sales) AS total_sales
FROM sales_data
GROUP BY Region
ORDER BY total_sales DESC
"""

pd.read_sql(query, conn)
```

The above query identifies which region generated the highest sales : West.

### Top 3 categories with highest avg_sales

```python
query = """
SELECT Category, AVG(Sales) AS avg_sales
FROM sales_data
GROUP BY Category
ORDER BY avg_sales DESC
"""

pd.read_sql(query, conn)
```

This query shows the average sales value for each product category.

### Top 3 categories with highest avg_sales

```python
query = """
SELECT Category,
       SUM(Profit) AS total_profit
FROM sales_data
GROUP BY Category
ORDER BY total_profit DESC
"""

pd.read_sql(query, conn)
```

This query identifies which product category generated the highest overall profit.

### MIN and MAX Profit

```python
query = """
SELECT MIN(Sales) AS min_sales, MAX(Sales) AS max_sales
FROM sales_data
"""

pd.read_sql(query, conn)
```

This query shows the lowest and highest sales values.

### MIN and MAX Profit

```python
query = """
SELECT MIN(Profit) AS min_profit, MAX(Profit) AS max_profit
FROM sales_data
"""

pd.read_sql(query, conn)
```

This query identifies the maximum profit earned and the maximum loss incurred in a single transaction.

## 5.Sort and limit results (top products, top categories)

### Top 10 Products by Sales

```python
query = """
SELECT "Product Name", Sales , Profit
FROM sales_data
ORDER BY Sales DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

This query identifies the products with the highest sales value in the dataset.

### Top Categories by Total Sales

## 6.Solve use cases (trends, top customers, duplicates)

### Top 10 Customers by Total Sales

```python
query = """
SELECT "Customer Name" , SUM(Sales) AS total_sales
FROM sales_data
GROUP BY "Customer Name"
ORDER BY total_sales DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

This query identifies the customers who contributed the highest total sales.

### Most Profitable Customers

```python
query = """
SELECT
    "Customer Name", SUM(Profit) AS total_profit
FROM sales_data
GROUP BY "Customer Name"
ORDER BY total_profit DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

This query identifies the customers contributing the highest overall profit.

### Yearly Sales Trend

```python
query = """
SELECT
    SUBSTR("Order Date",-4) AS year,
    SUM(Sales) AS total_sales
FROM sales_data
GROUP BY year
ORDER BY year
"""

pd.read_sql(query, conn)
```

This query shows how sales changed across different years, to identify seasonal trends.

### Duplicate Records Check

```python
query = """
SELECT  "Row ID" ,  COUNT(*) AS occurs
FROM sales_data
GROUP BY "Row ID"
HAVING COUNT(*) > 1
"""

pd.read_sql(query, conn)
```

This query identifies any Row IDs appearing more than once in the dataset.


### Check Duplicate Order IDs

```python
query = """
SELECT
    "Order ID",  COUNT(*) AS items
FROM sales_data
GROUP BY "Order ID"
HAVING COUNT(*) > 1
ORDER BY items DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

This query identifies orders that contain multiple items. Multiple occurrences of the same Order ID are expected because a single order can include more than one product.

### CASE on Profit

```python
query = """
SELECT
CASE
    WHEN Profit > 0 THEN 'Profit'
    ELSE 'Loss'
END AS order_status,
COUNT(*) AS total_orders
FROM sales_data
GROUP BY order_status
"""

pd.read_sql(query, conn)
```

This query classifies orders into profitable and loss-making orders and shows their distribution.

### Compare Sales using different Shipping Modes

```python
query = """
SELECT
    "Ship Mode",
    COUNT(*) AS total_orders,
    ROUND(SUM(Sales),2) AS total_sales
FROM sales_data
GROUP BY "Ship Mode"
ORDER BY total_sales DESC
"""

pd.read_sql(query, conn)
```

This query compares different shipping modes based on the number of orders and total sales generated.
